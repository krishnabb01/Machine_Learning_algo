```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_data = load_breast_cancer()
```

```python
cancer_data.keys()
```

```python
df = pd.DataFrame(cancer_data["data"],columns=cancer_data["feature_names"])
```

```python
df.head()
```

here we have multiple(30) features it would be usful for model
so dont use feature selection perhalf we use DIMENSIONALY REDUCTION.

```python
# Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
s_df = scaler.transform(df)
```

```python
s_df
```

```python
# Applying PCA Algorithms
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
pca_data = pca.fit_transform(s_df)
```

```python
pca_data
```

```python
pca.explained_variance_
```
