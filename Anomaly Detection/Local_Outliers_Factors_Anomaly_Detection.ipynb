```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

**Data**

```python
x, y = make_circles(n_samples=750,factor=0.3,noise=0.12)
```

**Visualize the Data**

```python
plt.scatter(x[:,0],x[:,1])
```

**Findout Outliers**

```python
from sklearn.neighbors import LocalOutlierFactor
loc = LocalOutlierFactor(n_neighbors=10)
```

```python
prdi = loc.fit_predict(x)
```

```python
index = np.where(prdi < 0)
index
```

**Visualize The Outliers**

```python
plt.scatter(x[:,0],x[:,1])
plt.scatter(x[index,0],x[index,1],edgecolors="r")
```
