Remove unwanted columns

```python
df.drop(columns=["Age","Married","Id_Factor","Owns_House"],inplace=True)
```

```python
# Spliting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("Defaulted",axis=1), df["Defaulted"], test_size=0.3)
```

```python
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(class_weight='balanced')
```

```python
log_reg.fit(X_train_scaled,y_train)
```

```python
y_pred = log_reg.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(f"Score : {accuracy_score(y_test,y_pred)}")
print(f"Report : {classification_report(y_test,y_pred)}")
```

### Hyper parameter

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear', 'saga'],
    'max_iter': [100, 200, 300],
    'fit_intercept': [True, False]
}
```

```python
cv = GridSearchCV(estimator=LogisticRegression(class_weight='balanced'),param_grid=parameter,cv=5,scoring="roc_auc")
cv.fit(X_train_scaled,y_train)
```

```python
cv.best_estimator_
```

```python
y_predi = cv.predict(X_test)
```

```python
print(f"Score : {accuracy_score(y_test,y_predi)}")
print(f"Report : {classification_report(y_test,y_predi)}")
```
