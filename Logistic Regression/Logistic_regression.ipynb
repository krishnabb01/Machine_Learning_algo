Remove unwanted columns

```python
df.drop(columns=["Age","Married","Id_Factor","Owns_House"],inplace=True)
```

```python
# Spliting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop("Defaulted",axis=1), df["Defaulted"], test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
```

```python
log_reg.fit(X_train_scaled,y_train)
```

```python
y_pred = log_reg.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(f"Score : {accuracy_score(y_test,y_pred)}")
Report = classification_report(y_test,y_pred,output_dict=True)
```

```python
classification_report_df = pd.DataFrame(Report).transpose()
classification_report_df
```

```python
log_reg.predict(scaler.transform([[35000,40000,800,0,3]]))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
conf_matrix_v2 = confusion_matrix(y_test,y_pred)
conf_matrix_v2
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_v2, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix for Logistic Regression Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
